## Data Wrangling

```python
# packages for data collection
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
#installing pandas_ply package
!pip install "pandas_ply"
#pandas-ply

from pandas_ply import install_ply, X
install_ply(pd)
```

```python
# a function to import the csv files easier
def read_file(csv):
    return pd.read_csv(csv)

# a function that will select the specific columns that I want 
def primary_player_stats(df):
    #select the primary stats
    df = df.ply_select('Season','G','Tm','Pos','MP','FG%','3P%','2P%','FT%','TRB','AST','STL','PTS')
    #renaming columns in snakecase for cleanliness
    df = (df
         .rename(columns = {'Season':'season'})
         .rename(columns = {'G':'games_played'})
         .rename(columns = {'Tm': 'team'})
         .rename(columns = {'Pos':'position'})
         .rename(columns = {'MP':'minutes_played'})
         .rename(columns = {'FG%':'field_goal_pct'})
         .rename(columns = {'FT%':'free_throw_pct'})
         .rename(columns = {'3P%':'three_point_pct'})
         .rename(columns = {'2P%':'two_point_pct'})
         .rename(columns = {'TRB':'total_rebounds'})
         .rename(columns = {'AST':'assists'})
         .rename(columns = {'STL':'steals'})
         .rename(columns = {'PTS':'points'})
         )
    return df

# a function that will create a dataframe comparing two arrays containing the player's average stats
def avg_compare(tab1,tab2):
    df = pd.DataFrame({
        "before_injury": tab1.mean(),
        "after_injury": tab2.mean(),
        "difference": tab2.mean() - tab1.mean(),
    })
    return df
```

## Players who have had ACL tears
- Al Jefferson (2009)
- David West (2011)
- Ricky Rubio (2012)


```python
al_jefferson = read_file('al_jefferson.csv')
al_jefferson = primary_player_stats(al_jefferson)
al_jefferson
```

```python
al_jefferson_before_injury = al_jefferson.take(np.arange(0,5))
al_jefferson_after_injury = al_jefferson.take(np.arange(5,14))

al_jefferson_average_comparison = avg_compare(al_jefferson_before_injury,al_jefferson_after_injury)
al_jefferson_average_comparison
```

```python
david_west = read_file('david_west.csv')
david_west = primary_player_stats(david_west)
david_west
```

```python
david_west_before_injury = david_west.take(np.arange(0,8))
david_west_after_injury = david_west.take(np.arange(8,15))

david_west_average_comparison = avg_compare(david_west_before_injury, david_west_after_injury)
david_west_average_comparison
```

```python
ricky_rubio = read_file('ricky_rubio.csv')
ricky_rubio = primary_player_stats(ricky_rubio)
ricky_rubio
```

```python
ricky_rubio_before_injury = ricky_rubio.take(np.arange(0,1))
ricky_rubio_after_injury = ricky_rubio.take(np.arange(1,8))

ricky_rubio_average_comparison = avg_compare(ricky_rubio_before_injury, ricky_rubio_after_injury)
ricky_rubio_average_comparison
```

## Players who have had Achilles ruptures
- Kobe Bryant (2013)
- Demarcus Cousins (2018)
- Wesley Matthews (2015)

```python
kobe_bryant = read_file('kobe_bryant.csv')
kobe_bryant = primary_player_stats(kobe_bryant)
kobe_bryant
```

```python
kobe_bryant_before_injury = kobe_bryant.take(np.arange(0,18))
kobe_bryant_after_injury = kobe_bryant.take(np.arange(18,20))

kobe_bryant_average_comparison = avg_compare(kobe_bryant_before_injury, kobe_bryant_after_injury)
kobe_bryant_average_comparison
```

```python
demarcus_cousins = read_file('demarcus_cousins.csv')
demarcus_cousins = primary_player_stats(demarcus_cousins)

#removes combined stats when player was traded to other teams
demarcus_cousins = demarcus_cousins.ply_where(X.team != 'TOT')

#re-index to clean up the index gaps
demarcus_cousins = demarcus_cousins.reset_index(drop=True)
demarcus_cousins
```

```python
demarcus_cousins_before_injury = demarcus_cousins.take(np.arange(0,9))
demarcus_cousins_after_injury = demarcus_cousins.take(np.arange(9,10))

demarcus_cousins_average_comparison = avg_compare(demarcus_cousins_before_injury, demarcus_cousins_after_injury)
demarcus_cousins_before_injury
```

```python
wesley_matthews = read_file('wesley_matthews.csv')
wesley_matthews = primary_player_stats(wesley_matthews)

#removes combined stats when player was traded to other teams
wesley_matthews = wesley_matthews.ply_where(X.team != 'TOT')

#re-index to clean up the index gaps
wesley_matthews = wesley_matthews.reset_index(drop = True)
wesley_matthews
```

```python
wesley_matthews_before_injury = wesley_matthews.take(np.arange(0,6))
wesley_matthews_after_injury = wesley_matthews.take(np.arange(6,12))

wesley_matthews_average_comparison = avg_compare(wesley_matthews_before_injury, wesley_matthews_after_injury)
wesley_matthews_average_comparison
```

### All players difference in averages before and after injury

```python
all_players = pd.DataFrame({
    'Jefferson': al_jefferson_average_comparison['difference'],
    'West': david_west_average_comparison['difference'],
    'Rubio': ricky_rubio_average_comparison['difference'],
    'Bryant': kobe_bryant_average_comparison['difference'],
    'Cousins': demarcus_cousins_average_comparison['difference'],
    'Matthews': wesley_matthews_average_comparison['difference'],
    })

#transpose data to compare horizontally and have variables at the columns
all_players = all_players.transpose()
all_players
```

## Data Visualizations 

The visualizations are a series of line plots that compare the difference of player's scoring statistics before and after their individual injuries. The positive numbers represent a player who does better after their injury and the negative numbers represent a player who does worse after their injury.

```python
plt.plot(all_players.points)
plt.xlabel('Players')
plt.ylabel('Avg Points Differences')
plt.title('Avg Difference in Points Between Players')
plt.show()
```

```python
plt.plot(all_players.minutes_played)
plt.xlabel('Players')
plt.ylabel('Avg Minutes Played Differences')
plt.title('Avg Difference in Minutes Played Between Players')
plt.show()
```

```python
plt.plot(all_players.three_point_pct)
plt.xlabel('Players')
plt.ylabel('Avg Three Point Differences')
plt.title('Avg Difference in Three Point Percentage Between Players')
plt.show()
```
